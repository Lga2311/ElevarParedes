```python
Paredes_df = pd.read_csv('paredes.csv', sep=';')
Paredes_df  = Paredes_df.dropna(how='all', axis=0)
Paredes_df  = Paredes_df.dropna(how='all', axis=1)
Paredes_df  = Paredes_df.drop('DeltaY', axis=1)
display(Paredes_df)
```

```python
# Paredes
paredes = {}
for par in Paredes_df['ID par']:
  paredes[par] = Paredes_df.loc[Paredes_df['ID par'] == par]
  paredes[par] = paredes[par].reset_index()

def carregar_paredes():
  return paredes

def id_pars():
  pars = []
  for i in carregar_paredes().keys():
    pars.append(i)
    
  return pars

id_pars()[0]
```

```python
Esquadrias_df = pd.read_csv('esquadrias.csv', sep=';')
Esquadrias_df = Esquadrias_df.dropna(how='all', axis=0)
Esquadrias_df = Esquadrias_df.dropna(how='all', axis=1)
#//Esquadrias_df = Esquadrias_df.drop('Unnamed: 12', axis=1)
display(Esquadrias_df)
```

```python
esquadrias = {}
for par in Paredes_df['ID par']:
  esquadrias[par] = Esquadrias_df.loc[Esquadrias_df['ID par'] == par]
  esquadrias[par] = esquadrias[par].reset_index()

display(esquadrias['P0004'])
```

```python

parede_selecionada = 'P0004'
esqBD = esquadrias
esqBD = esqBD[parede_selecionada] 

for i in range(len(esqBD)):
  xi = esqBD.loc[i, 'XiDelta']
  yi = esqBD.loc[i, 'YiDelta']
  larg = esqBD.loc[i,'Largura']
  alt = esqBD.loc[i, 'Altura']
  peitoril = esqBD.loc[i, 'Peitoril']
  peitoril = esqBD.loc[i, 'Peitoril']
  print(f'x={xi}, y={yi}, larg={larg}, alt={alt}, peitoril={peitoril}')
```
