```python
import pandas as pd
```

```python
# carrega arquivo
xls = pd.ExcelFile("Coletar Modelos de Paredes.xlsm")

# 'extrai' Paredes
Paredes_df = pd.read_excel(xls, sheet_name='Paredes')
Paredes_df  = Paredes_df.dropna(how='all', axis=0)
Paredes_df  = Paredes_df.dropna(how='all', axis=1)
Paredes_df  = Paredes_df.drop('DeltaY', axis=1)

display(Paredes_df)
```

```python
# 'extrai' ΔY
DeltaY = pd.read_excel(xls, sheet_name='Paredes')
DeltaY = DeltaY.loc[0, 'DeltaY']
display(DeltaY)
```

```python
# 'extrai' Esquadrias
Esquadrias_df = pd.read_excel(xls, sheet_name='Esq').dropna(how='all', axis=0)
display(Esquadrias_df)
```

```python
# 'extrai' LinFiadas
Fiadas_df = pd.read_excel(xls, sheet_name='Fiadas')
Fiadas_df = Fiadas_df.dropna(how='all', axis=0)
Fiadas_df = Fiadas_df.dropna(how='all', axis=1)
Fiadas_df = Fiadas_df.drop('Unnamed: 12', axis=1)

display(Fiadas_df)
```

```python
# 'extrai' Blocos
Blocos_df = pd.read_excel(xls, sheet_name='Blocos')
Blocos_df = Blocos_df.dropna(how='all', axis=0)
Blocos_df = Blocos_df.dropna(how='all', axis=1)
Blocos_df = Blocos_df.drop(['Unnamed: 13', 'Unnamed: 14'], axis=1)
display(Blocos_df)
```

```python
paredes = {}

for par in Paredes_df['ID par']:
  paredes[par] = Paredes_df.loc[Paredes_df['ID par'] == par]

print(paredes['P0010'])
```

```python
esquadrias = {}

for par in Esquadrias_df['ID par']:
  esquadrias[par] = Esquadrias_df.loc[Esquadrias_df['ID par'] == par]

print(esquadrias['P0010'])
```

```python
fiadas = {}

for par in Fiadas_df['ID par']:
  fiadas[par] = Fiadas_df.loc[Fiadas_df['ID par'] == par]

print(fiadas['P0010'])
```

```python
blocos = {}

for par in Blocos_df['ID par']:
  blocos[par] = Blocos_df.loc[Blocos_df['ID par'] == par]
  blocos[par] = blocos[par].reset_index()

display(blocos['P0010'])
```

```python
#Explicação: Criando funções para carregar as tabelas/dicionarios
# Paredes
def carregar_paredes():
  return paredes

# esquadrias
def carregar_esquadrias():
  return esquadrias

# fiadas
def carregar_fiadas():
  return fiadas

# blocos
def carregar_blocos():
  return blocos
```
