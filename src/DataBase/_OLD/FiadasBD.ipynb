```python
import pandas as pd
```

```python
# carrega arquivo
xls = pd.ExcelFile("Coletar Modelos de Paredes.xlsm")

# 'extrai' Paredes
Paredes_df = pd.read_excel(xls, sheet_name='Paredes')
Paredes_df  = Paredes_df.dropna(how='all', axis=0)
Paredes_df  = Paredes_df.dropna(how='all', axis=1)
Paredes_df  = Paredes_df.drop('DeltaY', axis=1)

display(Paredes_df)
```

```python
# 'extrai' ΔY
DeltaY = pd.read_excel(xls, sheet_name='Paredes')
DeltaY = DeltaY.loc[0, 'DeltaY']
display(DeltaY)
```

```python
# 'extrai' Esquadrias
Esquadrias_df = pd.read_excel(xls, sheet_name='Esq').dropna(how='all', axis=0)
display(Esquadrias_df)
```

```python
# 'extrai' LinFiadas
LinFiadas_df = pd.read_excel(xls, sheet_name='Fiadas')
LinFiadas_df = LinFiadas_df.dropna(how='all', axis=0)
LinFiadas_df = LinFiadas_df.drop('Unnamed: 10', axis=1)

display(LinFiadas_df)
```

```python
# 'extrai' Blocos
Blocos_df = pd.read_excel(xls, sheet_name='Blocos')
Blocos_df = Blocos_df.dropna(how='all', axis=0)
Blocos_df = Blocos_df.dropna(how='all', axis=1)
Blocos_df = Blocos_df.drop(['Unnamed: 11', 'Unnamed: 12'], axis=1)
display(Blocos_df)
```

# **Procurando Dados da Parede Selecionada**

```python
ParSelc = 'PAREDE 4 (7x)'
Pesq = True
i = 0

while Pesq:
  if Paredes_df.loc[i, 'Texto'] == ParSelc:
    ParSelcID = Paredes_df.loc[i, 'ID par']
    ParSelcXi = Paredes_df.loc[i, 'Xi']
    ParSelcYi = Paredes_df.loc[i, 'Yi']
    Pesq = False
  else:
    i += 1

  
print(ParSelcID)
print(ParSelcXi)
print(ParSelcYi)
```

```python
EsqSelc = Esquadrias.loc[Esquadrias['ID par'] == ParSelcID]
EsqSelc['Xid'] = round((Esquadrias['Xi'] - ParSelcXi) * 100, 0)
EsqSelc['Yid'] = round(Esquadrias['Peitoril'], 0)
display(EsqSelc)
```

```python
FiadaSelc = Fiadas_df.loc[Fiadas_df['ID par'] == ParSelcID]
FiadaSelc['Xid'] = round((Fiadas_df['Xi'] - ParSelcXi) * 100, 0)
FiadaSelc['Yid'] = round((Fiadas_df['Yi'] - ParSelcYi - DeltaY) * 100 - 1, 0)
display(FiadaSelc)
```
