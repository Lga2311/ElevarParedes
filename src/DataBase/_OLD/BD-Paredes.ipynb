```python
# carrega arquivo
xls = pd.ExcelFile("BD_MODELOS.xlsx")

# 'extrai' aba 1
paredesBD = pd.read_excel(xls, sheet_name='Paredes')
blocosBD = pd.read_excel(xls, sheet_name='Blocos')

display(blocosBD)
```

```python
#Visualizando dados da plotagem
PTX = []
PTY = []
TIPOS = []
FIADA = []

Blocos = {
  'FIADA': FIADA,
  'TIPOS': TIPOS,
  'X': PTX,
  'Y': PTY,
}

nF = 1
X = 0
Y = 0

for i in range(len(blocosBD)):
  tipo = blocosBD.loc[i, "Blocos"].replace("B", "")
  fiada = blocosBD.loc[i, "Fiada"]
    
  if fiada == nF:
    if i == 0:
        X = 0
    else:
        X += int(tipo) + 1
  else:
    X = 0
    Y += 20
    nF += 1
    
  PTX.append(X)
  PTY.append(Y)
  TIPOS.append(tipo)
  FIADA.append(fiada)
    
ParedeBlocos = pd.DataFrame(Blocos) 
display(ParedeBlocos.head(20))
  
```

```python
PTX = []
PTY = []
TIPOS = []
FIADA = []

Blocos = {
  'FIADA': FIADA,
  'TIPOS': TIPOS,
  'X': PTX,
  'Y': PTY,
}

nF = 1
X = 0
Y = 0

for i in range(len(blocosBD)):
  tipo = blocosBD.loc[i, "Blocos"].replace("B", "")
  fiada = blocosBD.loc[i, "Fiada"]
    
  if fiada == nF:
    X += int(tipo) + 1
  else:
    X = 200
    nF += 1
    Y += int(fiada) * 20
    
  PTX.append(X)
  PTY.append(Y)
  TIPOS.append(tipo)
  FIADA.append(fiada)
    
ParedeBlocos = pd.DataFrame(Blocos) 
display(ParedeBlocos.head(10))
```
